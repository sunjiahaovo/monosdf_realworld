```python
import numpy as np
import matplotlib.pyplot as plt
import pylab

plt.rc('font',family='Times New Roman')
# plt.rcParams['font.sans-serif'] = ['SimSun']  # 用来正常显示中文

plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['pdf.fonttype'] = 42

# pylab.subplots_adjust(wspace=0.36)
path = "../exps2/l435_rgbd_grids_7/2023_05_10_18_48_22/plots/logs/loss.txt"


data = np.loadtxt(path)
print(data.shape)
data = data[::10,:]

# print(data[:,7])
```

```python
from matplotlib.pyplot import title
data = data[0:6000,:]

fig = plt.figure(dpi=300,figsize=(16,12))

ax = fig.add_subplot(341)
ax.plot(data[:,0],data[:,3])
ax.grid(axis='y')
ax.set_title("loss")


ax = fig.add_subplot(342)
ax.plot(data[:,0],data[:,4])
ax.grid(axis='y')
ax.set_title("rgb_loss")

ax = fig.add_subplot(343)
ax.plot(data[:,0],data[:,5])
ax.grid(axis='y')
ax.set_title("eikonal_loss")

ax = fig.add_subplot(344)
ax.plot(data[:,0],data[:,6])
ax.grid(axis='y')
ax.set_title("smooth_loss")

ax = fig.add_subplot(345)
# ax.plot(data[0:,0],data[0:,7])
ax.grid(axis='y')
ax.set_title("depth_loss")

ax = fig.add_subplot(346)
ax.plot(data[:,0],data[:,8])
ax.grid(axis='y')
ax.set_title("normal_l1")


ax = fig.add_subplot(347)
ax.plot(data[:,0],data[:,9])
ax.grid(axis='y')
ax.set_title("normal_cos")

ax = fig.add_subplot(348)
ax.plot(data[:,0],data[:,2])
ax.grid(axis='y')
ax.set_title("psnr")

ax = fig.add_subplot(349)
ax.plot(data[:500,0],data[:500,10])
ax.grid(axis='y')
ax.set_title("pose")

ax = fig.add_subplot(3,4,10)
ax.plot(data[10:,0],np.log(data[10:,11]))
ax.grid(axis='y')
ax.set_title("uncer")


ax = fig.add_subplot(3,4,11)
ax.scatter(data[10:,2],np.log(data[10:,11]))
# ax.grid(axis='y')
ax.set_title("uncer-psnr")


fig.savefig("conver.jpg",dpi=300, bbox_inches = 'tight')

plt.show()
```
